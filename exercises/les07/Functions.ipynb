## Functions

```python
#Doel: uitrekenen van 10^3

def power(x, e=2): #e=2 is een default parameter
    result = x**e
    return result

a = 10
b = 3
res1 = power(a, b) #positionele argumenten
res2 = power(b, a) #volgorde maakt uit, antwoord is fout

print(res1)
print(res2)

# keyword argumenten
res3 = power(x=a, e=b) #function call
res4 = power(e=b, x=a)
print(res3)
print(res4)

RecursionError: maximum recursion depth exceeded while calling a Python object
```

# Scoping

Voordelen van scoping:
- variabelen namen kunnen op meerdere plekken gebruikt worden
- functies zijn meer afgesloten

## Local scope

```python
def calc_sum(x, y):
    result = x + y #result is "gescoped". Local scope
    print('calc_sum', result)
    return result 
    
def calc_multiply(x, y):
    result = x * y #result is "gescoped"
    print('calc_multiply', result)
    return result

#print(result) #NameError: name 'result' is not defined
print(calc_sum(2, 3))
print(calc_multiply(2, 3))

result = "bla" #global scope
print(result)
```

Conclusie: functie hebben een local scope. Deze is niet zichtbaar buiten de functie.  
Zie argumenten/parameters als deur naar binnen.  
Return is de deur naar buiten.  
Wat binnen de kamer is, blijft binnen de kamer.

Dus: local scope is niet te zien in de global scope.
